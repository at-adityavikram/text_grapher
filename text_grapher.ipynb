## Constructing a Blank Graph

A blank graph is just a grid of characters. I used periods because they don't have a large visual footprint, but they allow for easy debugging. 

Here I use a list comprehension to construct a 2D array of periods:

```python
graph = [
    ['.' for x in range(60)]
    for y in range(30) 
    ]
```

Working with the grid as a 2D array will make it easy to replace characters at given coordinates. But a way to preview the graph is required. 

This function joins the characters in each row, and then joins the rows with new line characters and then prints out the resulting string:

```python
def render(graph) -> str:
    print('\n'.join([''.join(row) for row in graph]))

render(graph)
```

Now "drawing" on the grid happens by simply replacing the periods with other characters. The following code draws a diagonal line on the graph by replacing the periods with "M"

```python
for i in range(5, 55):
    graph[i//2][i] = 'M'

render(graph)
```
